```python
import time
import os
from glob import glob
from pathlib import Path
import math
import numpy as np
import random


def run_os_listdir(directory):
    a = time.time_ns()
    dirname = Path(directory)
    fu = [os.path.join(directory, o) for o in os.listdir(directory) if os.path.isdir(os.path.join(directory, o))]    
    return fu

directory =r"/root//preprocessed/all_preproceeed_codes_by_url"

fu = run_os_listdir(directory)
all_rep_nums = 0
for d in fu:
    sfu = run_os_listdir(d)
    for sd in sfu:
        all_rep_nums +=1
```

```python
fu
```

```python
all_rep_nums
```

```python
train_pre = 80
train_rep_nums = math.ceil(train_pre*all_rep_nums/100)
test_rep_nums = all_rep_nums-train_rep_nums
train_rep = []
test_rep = []
```

```python
fu
```

```python
for d in fu:
    sfu = run_os_listdir(d)
    if(len(sfu)==0):
        continue
    print(sfu)    
    selected_test = math.ceil(test_rep_nums*(len(sfu)/all_rep_nums))
    all_list = np.arange(len(sfu))
    test = random.sample(range(0, len(sfu)), selected_test)
    train = np.delete(all_list, test)
    for d in test:
        test_rep.append(sfu[d])
    for d in train:
        train_rep.append(sfu[d])
```
